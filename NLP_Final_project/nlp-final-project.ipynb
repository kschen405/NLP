```python
import re
import json
import contractions

for index in range(len(train_data)):
    premise_data = []
    for file_name in train_data[index]['metadata']['premise_articles'].values():
        article = []
        with open(f'articles/{file_name}') as f:
            tmp_article = json.load(f)
            for s in tmp_article:
                if len(s) != 1 and (s[0].isalnum() or s[0] == '@') and s.startswith('Link:') == False:
                    filter_s = re.sub(r'http\S+', '', s)
                    filter_s = re.sub(r"[^a-zA-Z0-9 .']", "", filter_s)
                    filter_s = contractions.fix(filter_s).lower() #
                    article.append(filter_s)
        premise_data += article

    json_object = json.dumps(premise_data, indent=4)
    with open(f"filtered_articles/{index}.json", "w") as outfile:
        outfile.write(json_object)

# print(premise_data)
```

```python
def count_labels(data):
    ratings = {}
    
    for item in data:
        rating = item['label']['rating']
        if rating in ratings:
            ratings[rating] += 1
        else:
            ratings[rating] = 1

    print(ratings)
```
